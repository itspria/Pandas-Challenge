```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#Get unique players and store the length in a variable
players = purchase_data['SN'].unique()
uniquePlayersCount = len(players)

print(f"Total number of players = {uniquePlayersCount}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueItems = purchase_data['Item ID'].unique()
itemCount = len(uniqueItems)
avgPrice = "${:,.2f}".format(purchase_data['Price'].mean())
purchase = purchase_data['Purchase ID'].unique()
purchaseCount =  len(purchase)
total = "${:,.2f}".format(purchase_data['Price'].sum())

#create a new dataframe with the values calculated above
summaryDF = pd.DataFrame({"Number of Unique Items" : [itemCount],
                         "Average Price" : avgPrice,
                         "Number of Purchases" : [purchaseCount],
                         "Total Revenue" : total})
summaryDF
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create a pivot table
genderTable = purchase_data.pivot_table(index = 'Gender', values=['SN'], aggfunc=lambda x: len(x.unique()))
genderTable = genderTable.rename(columns = { 'SN': 'Total Count'})

# calculating percentage
genderTable['Percentage of Players'] = genderTable['Total Count']/uniquePlayersCount*100

# add formatting
genderTable['Percentage of Players'] = genderTable['Percentage of Players'].map("{:,.2f}%".format)

genderTable = genderTable.sort_values('Total Count', ascending = False)
genderTable
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchaseSummary = purchase_data.groupby('Gender').agg(Purchase_Count = ('Purchase ID','count'),
                                                      Average_Purchase_Price = ('Price', np.mean), 
                                                      Total_Purchase_Value = ('Price', np.sum),
                                                      Count = ('SN', lambda x: len(x.unique())))

purchaseSummary["Avg_Total_Purchase_per_Person"] = purchaseSummary['Total_Purchase_Value']/purchaseSummary['Count']

# add formatting
purchaseSummary['Average_Purchase_Price'] = purchaseSummary['Average_Purchase_Price'].map("${:,.2f}".format)
purchaseSummary['Total_Purchase_Value'] = purchaseSummary['Total_Purchase_Value'].map("${:,.2f}".format)
purchaseSummary['Avg_Total_Purchase_per_Person'] = purchaseSummary['Avg_Total_Purchase_per_Person'].map("${:,.2f}".format)

# Delete the Count column as it is not required
del purchaseSummary['Count']

purchaseSummary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
ageAnalysis = purchase_data.copy()
#Create datafrane with the required bins
ageAnalysis['Age Groups'] = pd.cut(ageAnalysis['Age'], bins=[1,10,15,20,25,30,35,40,100], labels=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'], right=False)

ageDemographics = ageAnalysis.groupby('Age Groups').agg(Total_Count = ('SN',  lambda x: len(x.unique())))
ageDemographics['Percentage of Players'] = ageDemographics['Total_Count']/uniquePlayersCount*100
ageDemographics['Percentage of Players'] = ageDemographics['Percentage of Players'].map("{:,.2f}%".format)
ageDemographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
agewisePurchase = ageAnalysis.groupby('Age Groups').agg(Purchase_Count = ('Purchase ID','count'),
                                                      Average_Purchase_Price = ('Price', np.mean), 
                                                      Total_Purchase_Value = ('Price', np.sum),
                                                      Count = ('SN', lambda x: len(x.unique())))

agewisePurchase["Avg_Total_Purchase_per_Person"] = agewisePurchase['Total_Purchase_Value']/agewisePurchase['Count']

# add formatting
agewisePurchase['Average_Purchase_Price'] = agewisePurchase['Average_Purchase_Price'].map("${:,.2f}".format)
agewisePurchase['Total_Purchase_Value'] = agewisePurchase['Total_Purchase_Value'].map("${:,.2f}".format)
agewisePurchase['Avg_Total_Purchase_per_Person'] = agewisePurchase['Avg_Total_Purchase_per_Person'].map("${:,.2f}".format)

# Delete the Count column as it is not required
del agewisePurchase['Count']

agewisePurchase
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders = purchase_data.groupby('SN').agg(Purchase_Count = ('Purchase ID','count'),
                                                      Average_Purchase_Price = ('Price', np.mean), 
                                                      Total_Purchase_Value = ('Price', np.sum))

spenders = spenders.sort_values('Total_Purchase_Value', ascending = False)
spenders[['Average_Purchase_Price','Total_Purchase_Value']] = spenders[['Average_Purchase_Price','Total_Purchase_Value']].applymap("${:,.2f}".format)
spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popularItems = purchase_data[['Item ID', 'Item Name', 'Price']]
popularItemsDF = popularItems.groupby(['Item ID','Item Name']).agg(Purchase_Count = ('Item ID','count'), Item_Price = ('Price', np.mean))
popularItemsDF['Total_Purchase_Value'] = popularItemsDF['Purchase_Count'] * popularItemsDF['Item_Price']
popularItemsDFCopy = popularItemsDF.copy()

popularItemsDF[['Item_Price', 'Total_Purchase_Value']] = popularItemsDF[['Item_Price', 'Total_Purchase_Value']].applymap("${:,.2f}".format)

popularItemsDF = popularItemsDF.sort_values('Purchase_Count', ascending = False)
                                                                 
popularItemsDF.head(10)
                                                                  
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profitDF = popularItemsDFCopy.sort_values('Total_Purchase_Value', ascending = False)
profitDF[['Item_Price', 'Total_Purchase_Value']] = profitDF[['Item_Price', 'Total_Purchase_Value']].applymap("${:,.2f}".format)
profitDF.head()
```
